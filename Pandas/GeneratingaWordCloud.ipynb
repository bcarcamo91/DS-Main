```python
#Creating a word cloud with Python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample data
data = {'text': ["apple banana orange banana orange apple apple",
                 "orange apple banana banana banana",
                 "apple apple orange orange orange orange"]}

df = pd.DataFrame(data)

# Concatenate all the text from the 'text' column
text = ' '.join(df['text'])

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Plot the WordCloud image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
