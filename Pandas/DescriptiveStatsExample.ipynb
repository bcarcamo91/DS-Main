```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Age': [25, 30, 35, 28, 22, 40, 32, 38, 45, 27],
        'Salary': [50000, 60000, 70000, 55000, 48000, 75000, 67000, 80000, 90000, 52000],
        'Experience': [3, 5, 7, 4, 2, 9, 6, 8, 10, 3]}

df = pd.DataFrame(data)

# Display basic descriptive statistics
descriptive_stats = df.describe()

# Display additional statistics
median = df.median()
variance = df.var()
std_dev = df.std()

# Display results
print("Basic Descriptive Statistics:")
print(descriptive_stats)

print("\nMedian:")
print(median)

print("\nVariance:")
print(variance)

print("\nStandard Deviation:")
print(std_dev)
```

```python
# Create a box plot
plt.figure(figsize=(8, 6))
df.boxplot()
plt.title('Box Plot of Age, Salary, and Experience')
plt.ylabel('Values')
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Salary'])
plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
correlation_coefficient = df['Age'].corr(df['Salary'])
print(f"Correlation Coefficient between Age and Salary: {correlation_coefficient:.2f}")
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Salary'], df['Experience'])
plt.title('Scatter Plot of Salary vs Experience')
plt.xlabel('Salary')
plt.ylabel('Experience')
plt.show()
```

```python
# Calculate the correlation coefficient
correlation_coefficient = df['Salary'].corr(df['Experience'])
print(f"Correlation Coefficient between Salary and Experience: {correlation_coefficient:.2f}")
```
